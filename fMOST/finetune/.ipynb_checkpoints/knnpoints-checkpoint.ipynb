```python
import numpy as np
import pandas
from scipy.spatial import distance
```

```python
sub = pandas.read_csv('sub.csv', names=['x','y','z'])
tar = pandas.read_csv('tar.csv', names=['x','y','z'])
baseline = pandas.read_csv('baseline.csv', names=['x','y','z'])
```

```python
print(sub.shape, tar.shape, baseline.shape)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
knn = NearestNeighbors(n_neighbors=6)
```

```python
b = baseline.values

#b = b[1:-1,...]
b = b[1:1952,...]

print(b.shape, b)
```

```python
s = sub.values

s = s[1:63,...]

t = tar.values

t = t[1:63,...]

print(s.shape, t.shape)

test = np.concatenate((s,t))

print(test.shape)

print(test)
```

```python
knn.fit(s)
```

```python
from sklearn.neighbors import KDTree


result = np.concatenate((b,test))

print(result.shape)

tree = KDTree(result)

all_nn_indices = tree.query_radius(result, r=16)

all_nns = [ [result[idx] for idx in nn_indices] for nn_indices in all_nn_indices ]

for nns in all_nns:
    print(nns)
```

```python
for nn_indices in all_nn_indices:
    print(nn_indices)
```
